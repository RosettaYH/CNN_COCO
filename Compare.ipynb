## Food101 - food labels

```python
food_list = []
with open('/Users/rosetta/Desktop/food-101/meta/labels.txt', 'r') as txt:
      paths = [read.strip() for read in txt.readlines()]
      for p in paths:
        food_list.append(p)

print(food_list)
```

```python
for i in range(len(food_list)):
    if(food_list[i] == "Omelette"):
        food_list[i] = "Omelet"     
```

## Get the matching IDs 

```python
# Find the number of times 
def occur(appear, foodName, idNum, numWords):
    final = ""
    sum = 0

    for f in appear:
        # Find if foodName already exists 
        sum += f[0].count(foodName)
    if(sum >= numWords):
        # The set need to contain all the words in the Food101 label
        final = foodName
    return (final, idNum)
```

```python
IDList = []

for food in food_list:
    print(food)
    num = 0          # For getting FoodID from USDA dataset
    appear = []      # Store valid FoodNames and FoodIDs
    ID = []          # Store valid FoodIDs
    numWords = food.count(" ") + 1           # Number of words of the current Food101 label

    for foodName in foodNames:
        foodName = re.sub(r'([^\s\w]|_)+', '', foodName)      # Remove all non alphanumeric characters, except for space
        for f in foodName.split(" "):  
            # Loop through each word
            match = set(re.findall(f, food, re.IGNORECASE))       
            if(len(match) > 0 and num < 7793):
                for i in match:
                    # Loop through each matched word set
                    if(i.isalnum() and len(i) > 1): 
                        # Makes sure the word is alphanumeric and is not a single character
                        appear.append([foodName, foodID[num]])     
        num+=1
    for a in appear:
        o = occur(appear, a[0], a[1], numWords)
        if(len(o[0]) > 0):
            # Add ID if the all words of Food101 label occurs 
            ID.append(o[1])
    for i in ID:
        # Remove duplicates
        if(len(ID) > 1 and ID.count(i) >= 0):
            ID.remove(i)
        

    IDList.append(ID)

print(IDList)
```

## Get the IDs for NDB_number 

```python
NDBList = []

for id in IDList:

    ID = []
    if(len(id) != 0):
        for i in id:
            num = 0
            for ndb in NDB_number:   
                if(num < 7793 and i == fdc_id[num]):
                    ID.append(ndb)
                num+=1                
    NDBList.append(ID)
print(NDBList)
```

## Add all the Univeristy of Minnesota IDs

```python
MList = []
for food in food_list:
    ID = []
    if(food == "Ceviche"):
        ID.append(18934)
    elif(food == "Churros"):
        ID.append(19083)
    elif(food == "Guacamole"):
        ID.append(3309)
    elif(food == "Huevos rancheros"):
        ID.append(19258)
    elif(food == "Pad thai"):
        ID.append(111013)
        ID.append(111014)
    elif(food == "Red velvet cake"):
        ID.append(116446)
    elif(food == "Tiramisu"):
        ID.append(32864)
    MList.append(ID)
print(MList)
```

```python
foodList = {"Food": food_list, "fdc_id": IDList, "NDB_number": NDBList, "M_id": MList}
```
