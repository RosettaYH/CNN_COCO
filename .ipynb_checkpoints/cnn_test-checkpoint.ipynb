```python
from tensorflow.keras.models import load_model

model_best = load_model('./best_model_class50.hdf5', compile = False)
```

```python
from keras.preprocessing.image import load_img, img_to_array
import matplotlib.image as img
from tensorflow.keras.preprocessing import image
import numpy as np

food_list = []
with open('./food-101/meta/classes.txt', 'r') as txt:
  paths = [read.strip() for read in txt.readlines()]
  for p in paths:
    food_list.append(p)
print(food_list)
```

```python
import matplotlib.pyplot as plt

def predict_class(model, images, show = True):
  for img in images:
    img = image.load_img(img, target_size=(299, 299))
    img = image.img_to_array(img)                    
    img = np.expand_dims(img, axis=0)         
    img /= 255.                                      

    pred = model.predict(img)
    index = np.argmax(pred)
    food_list.sort()
    pred_value = food_list[index]
    pred_proba = pred[0][index]
    if show:
        plt.imshow(img[0])                           
        plt.axis('off')
        plt.suptitle(pred_value)
        plt.title(pred_proba)
        plt.show()
```

```python
images = []

# Change path
images.append('/Users/rosetta/Downloads/c1.jpg')
images.append('/Users/rosetta/Downloads/ap2.png')

images.append('/Users/rosetta/Downloads/fr1.jpg')    
images.append('/Users/rosetta/Downloads/cc2.jpg')

images.append('/Users/rosetta/Downloads/pd1.jpg')
images.append('/Users/rosetta/Downloads/rvc1.jpg')


predict_class(model_best, images, True)
```
